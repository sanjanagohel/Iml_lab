```python
#Gohel Sanjana D.
#216270307093
#Write a NumPy program to create a 3x3 matrix with values ranging from 2 to 10

import numpy as np

# Create an array of values ranging from 2 to 10
values = np.arange(2, 11)

# Reshape the array into a 3x3 matrix
matrix = values.reshape(3, 3)

# Print the resulting matrix
print(matrix)
```

```python
#Gohel Sanjana D.
#216270307093
##Write a NumPy program to append values at the end of an array

import numpy as np

# Create a NumPy array
original_array = np.array([100, 200, 300, 400, 500])

# Values to append
values_to_append = np.array([600, 700])

# Append the values to the end of the original array
new_array = np.append(original_array, values_to_append)

# Print the new array
print(new_array)
```

```python
#Gohel Sanjana D.
#216270307093
#Write a NumPy program to create another shape from an array without changing its data (3*2 to 2*3)


import numpy as np

# Create a 3x2 NumPy array
original_array = np.array([[1, 2],
                            [3, 4],
                            [5, 6]])

# Reshape the array to 2x3 without changing data
reshaped_array = original_array.reshape(2,3)

# Print the reshaped array
print(reshaped_array)
```
