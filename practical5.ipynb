```python
#216270307093
#Gohel Sanjana

#Write a NumPy program to implement following operation â€¢	to compute the mean, standard deviation, and variance of a given array along the second axis

import numpy as np

# Create a sample 2D array
array_2d = np.array([[12, 15, 20], [10, 25, 5], [30, 35, 40]])

# Compute mean, standard deviation, and variance along the second axis
mean_values = np.mean(array_2d, axis=1)
std_deviation = np.std(array_2d, axis=1)
variance = np.var(array_2d, axis=1)

print("Original array:\n", array_2d)
print("Mean along the second axis:", mean_values)
print("Standard deviation along the second axis:", std_deviation)
print("Variance along the second axis:", variance)
```
